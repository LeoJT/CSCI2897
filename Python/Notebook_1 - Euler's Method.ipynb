# Notebook_1 - Euler's Method

Numerical solutions and Euler's method

### Import needed packages

```python
import numpy as np
import matplotlib.pyplot as plt
# let's set things up so we have nice font sizes
plt.rcParams.update({'font.size': 12})
```

### Goals

We want to get a numerical solution to the ODE:
$$y' = 2xy$$
with the initial condition
$$y(1)=1$$

We're going to use Euler's method:
$$y_{n+1} = y_{n} + h f(x_n,y_n)$$
$$x_{n+1} = x_{n} + h$$

```python
# This function takes in where we are "now" (xn,yn) as well as a
# step size h, and then takes a step in the right direction. 
# Here, line 5 has "hard coded" our equation y'=2xy
def forward_euler_step(xn,yn,h):
    ynext = yn + h * 2 * xn * yn  #f=2xy
    xnext = xn + h
    return xnext,ynext
```

```python
# Let's start by choosing a step size
h=0.1

# Let's also decide on a final value for x. 
x_final = 3

# Let's initialize some empty Python lists to store the solution
x_solution = []
y_solution = []

# And let's put in the initial conditions: x=1 and y=1
# by appending them to the end of the lists.
x_solution.append(1)
y_solution.append(1)

# We'll keep stepping forward until we've reached x_final
while ( x_solution[-1]<x_final ):
    # take a step
    x_next,y_next=forward_euler_step(x_solution[-1],y_solution[-1],h)
    # take the new positions we just computed and put them 
    # at the end of the solution
    x_solution.append(x_next)
    y_solution.append(y_next)
```

For this problem, we actually happen to know the exact solution, which is $$y(x) = e^{x^2-1}$$
We can define this as a function:

```python
def true_solution(x):
    # e^(x^2-1), the true solution
    return np.exp(np.array(x)**2-1)
```

And we can get the true solution's `y` values for each of the `x` values that are in our numerical solution. 

```python
y_true = true_solution(x_solution)
```

### Plot

```python
# Set up the axes
fig, ax = plt.subplots(nrows=1,ncols=1)
# Plot the numerical solution
ax.plot(x_solution,y_solution,label='numerical soln')
# Plot the true solution
ax.plot(x_solution,y_true,label='true soln')

# Make the plot attractive
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend(loc='best')
ax.set_title('numerical solution example')
```

# Bonus

As we make the step size smaller and smaller, we'll typically get a better and better answer.  Notice how above, there's quite the gap between the numerical solution and the true solution? 

Here, we'll drop the step size down to 1/10 of what it was. See how the gap narrows?

```python
# Let's start by choosing a step size
h=0.01

# Let's also decide on a final value for x. 
x_final = 3

# Let's initialize some empty Python lists to store the solution
x_solution = []
y_solution = []

# And let's put in the initial conditions: x=1 and y=1
# by appending them to the end of the lists.
x_solution.append(1)
y_solution.append(1)

# We'll keep stepping forward until we've reached x_final
while ( x_solution[-1]<x_final ):
    # take a step
    x_next,y_next=forward_euler_step(x_solution[-1],y_solution[-1],h)
    # take the new positions we just computed and put them 
    # at the end of the solution
    x_solution.append(x_next)
    y_solution.append(y_next)
    
y_true = true_solution(x_solution)
```

```python
# Set up the axes
fig, ax = plt.subplots(nrows=1,ncols=1)
# Plot the numerical solution
ax.plot(x_solution,y_solution,label='numerical soln (h={})'.format(h))
# Plot the true solution
ax.plot(x_solution,y_true,label='true soln')

# Make the plot attractive
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend(loc='best')
ax.set_title('numerical solution example')
```
